```python
import keras_cv
from translate import Translator
import matplotlib.pyplot as plt
```

## (1) generate image

### - (1) 이미지 사이즈 (img_width=512, img_height=512) 40분 소요

```python
model = keras_cv.models.StableDiffusion(img_width=512, img_height=512)
translator = Translator(from_lang="ko", to_lang="en")

def generate_img(text:str, cnt:int) -> list:
    text = translator.translate(text)
    images = model.text_to_image(text, batch_size=cnt)
    return images

def plot_image(images) -> None:
    plt.figure(figsize=(20, 20))

    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i])
        plt.axis("off")
        plt.tight_layout()
        plt.savefig(f"text.png")

if __name__ == '__main__':
    images = generate_img("피아노 치는 고양이", 1)
    plot_image(images)
```

### - (2) 이미지 사이즈 줄임 (img_width=256, img_height=256) 10분 소요

```python
model = keras_cv.models.StableDiffusion(img_width=256, img_height=256)
translator = Translator(from_lang="ko", to_lang="en")

def generate_img(text:str, cnt:int) -> list:
    text = translator.translate(text)
    images = model.text_to_image(text, batch_size=cnt)
    return images

def plot_image(images) -> None:
    plt.figure(figsize=(20, 10))

    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i])
        plt.axis("off")
        plt.tight_layout()
        plt.savefig(f"text.png")

if __name__ == '__main__':
    images = generate_img("바이올린 켜는 강아지", 1)
    plot_image(images)
```
